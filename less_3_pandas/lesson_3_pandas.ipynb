# Question 1

```python
import pandas as pd
import numpy as np
```

```python

energy = pd.read_excel("Energy Indicators.xls", skiprows = 18, skipfooter = 38, usecols = 'C:F', header = None) 
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy['Energy Supply'] *= 1000000
energy[energy['Energy Supply per Capita'] == '...'] = np.nan
energy['Country'] = energy['Country'].replace({
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
    })
energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex = True)
energy['Country'] = energy['Country'].str.replace(r'\s * \(/*?\)\s' , '', regex = True)

GDP = pd.read_csv('world_bank.csv', skiprows = 4)
GDP.rename(columns = {'Country Name': 'Country'}, inplace = True)
GDP['Country'] = GDP['Country'].replace({
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
    })
    
ScimEn = pd.read_excel('scimagojr.xlsx')

def answer_one():
    
    df = pd.merge(pd.merge(energy, GDP, on = 'Country'), ScimEn, on = 'Country')
    df.set_index('Country', inplace = True)
    df = df[['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    df = (df.loc[df['Rank'].isin([i for i in range(1, 16)])])
    df.sort_values(['Rank'], inplace = True)
    return df

answer_one()
```

# Question 2

```python
def answer_two():
    Top15 = answer_one()
    avgGDP = Top15[['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']].apply(lambda x: np.mean(x),axis=1)
    avgGDP.sort_values(inplace = True, ascending = False)
    return avgGDP

answer_two()
```

# Question 3

```python
def answer_three():
    Top15 = answer_one()
    change_GDP = float(Top15.iloc[5]['2015'] - Top15.iloc[5]['2006'])
    return change_GDP

answer_three()
```

# Question 4

```python
def answer_four():
    Top15 = answer_one()
    Top15["Ratio"] = Top15["Self-citations"] / Top15["Citations"]
    return Top15["Ratio"].idxmax(), max(Top15["Ratio"])

answer_four()
```

# Question 5

```python
def answer_five():
    Top15 = answer_one()
    Top15["BronzMedal"] = Top15["Energy Supply"] / Top15["Energy Supply per Capita"]
    y = Top15["BronzMedal"].sort_values(ascending = False)
    return y.index[2]

answer_five()
```

# Question 6

```python
def answer_six():
    Top15 = answer_one()
    Top15["CDPP"] = Top15["Energy Supply"] / Top15["Energy Supply per Capita"]
    Top15['Citable documents per Capita'] = Top15['Citable documents'] / Top15['CDPP']
    correlation= (Top15['Citable documents per Capita'].astype(float).corr(Top15['Energy Supply per Capita']))
    return float(correlation)

answer_six()
```

# Question 7

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    groups = pd.DataFrame(columns = ['size', 'sum', 'mean', 'std'])
    Top15['Estimate Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    for group, frame in Top15.groupby(ContinentDict):
        groups.loc[group] = [len(frame), frame['Estimate Population'].sum(), frame['Estimate Population'].mean(), frame['Estimate Population'].std()]
    return groups

answer_seven()
```

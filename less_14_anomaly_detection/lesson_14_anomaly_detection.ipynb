```python
from sklearn.covariance import EllipticEnvelope
from sklearn.datasets import make_blobs, make_moons
from sklearn.ensemble import IsolationForest
from sklearn.kernel_approximation import Nystroem
from sklearn.linear_model import SGDOneClassSVM
from sklearn.neighbors import LocalOutlierFactor
from sklearn.pipeline import make_pipeline
```

```python
X = df[['V1', 'V2']]
y = df[['Class']]
```

```python
np.random.seed(42)
sample_idx = np.random.choice(len(X), size = 5000, replace = False)
X_sample = X.iloc[sample_idx].values
y_sample = y.iloc[sample_idx].values
```

```python
anomaly_algorithms = {
    'Isolation Forest': IsolationForest(contamination = 0.01, random_state = 42),
    'Local Outlier Factor': LocalOutlierFactor(n_neighbors = 20, contamination = 0.01, novelty = True),
    'One-Class SVM (SDG)': make_pipeline(Nystroem(gamma = 0.2, random_state = 1), SGDOneClassSVM(nu = 0.01, shuffle = True, random_state = 42)),
    'Elliptic Envelope': EllipticEnvelope(contamination = 0.01, random_state = 42)
}
```

```python
plt.figure(figsize = (15, 12))
for i, (name, model) in enumerate(anomaly_algorithms.items(), 1):
    try:
        model.fit(X_sample)
        y_pred = model.predict(X_sample)
        y_pred = np.where(y_pred == -1, 1, 0)

        plt.subplot(3, 2, i)
        plt.scatter(X_sample[:,0], X_sample[:, 1], c = y_pred, cmap = 'coolwarm', s = 10)
        plt.title(name)
        plt.xlabel('V1')
        plt.ylabel('V2')
    except Exception as e:
        print(f"Error in {name}: {e}")

plt.tight_layout()
plt.show()
```

```python
def IQR_method(df,n,features, k=1.5):
    outlier_list = []

    for column in features:

        Q1 = np.percentile(df[column], 25)
        Q3 = np.percentile(df[column],75)

        IQR = Q3 - Q1

        outlier_step = k * IQR

        outlier_list_column = df[(df[column] < Q1 - outlier_step) | (df[column] > Q3 + outlier_step )].index

        outlier_list.extend(outlier_list_column)

    outlier_list = Counter(outlier_list)
    multiple_outliers = list( k for k, v in outlier_list.items() if v > n )

    df1 = df[df[column] < Q1 - outlier_step]
    df2 = df[df[column] > Q3 + outlier_step]

    print('Total number of outliers is:', df1.shape[0]+df2.shape[0])

    return multiple_outliers
```

```python
from sklearn.metrics import classification_report, confusion_matrix

outlier_indices = IQR_method(df, n=1, features=features, k=4.5)

df_out = df[features].copy()
df_out['anomaly'] = 'normal'
df_out.loc[outlier_indices, 'anomaly'] = 'outlier'
```

```python
df_out
```

```python
df_norm = df_out[df_out['anomaly'] == 'normal']
df_outlier  = df_out[df_out['anomaly'] == 'outlier']

fig1, axes1 = plt.subplots(nrows=3, ncols=2, figsize=(12, 8))
fig1.suptitle('Feature Distributions: Normal Observations', size=16)

for ax, feat in zip(axes1.flatten(), features):
    ax.hist(
        df_norm[feat],
        bins=60,
        alpha=0.7,
        edgecolor='white'
    )
    ax.set_title(f"{feat} (normal)")
    ax.set_xlabel(feat)
    ax.set_ylabel("Count")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()


fig2, axes2 = plt.subplots(nrows=3, ncols=2, figsize=(12, 8))
fig2.suptitle('Feature Distributions: IQR Outliers', size=16)

for ax, feat in zip(axes2.flatten(), features):
    ax.hist(
        df_outlier[feat],
        bins=60,
        alpha=0.7,
        edgecolor='white'
    )
    ax.set_title(f"{feat} (outlier)")
    ax.set_xlabel(feat)
    ax.set_ylabel("Count")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
df_out['pred_iqr'] = 0
df_out.loc[outlier_indices, 'pred_iqr'] = 1

y_true = df['Class']
y_pred = df_out['pred_iqr']

print("=== IQR Outlier Detection vs. True Fraud Labels ===")
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=['non‑fraud','fraud']))

cm = confusion_matrix(y_true, y_pred)
cm_df_out = pd.DataFrame(cm,
                     index=['true_non‑fraud','true_fraud'],
                     columns=['pred_non‑fraud','pred_fraud'])
print("\nConfusion Matrix:")
print(cm_df_out)
```

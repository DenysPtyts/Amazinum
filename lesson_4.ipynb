```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C:/amazinum/Car_Insurance_Claim.csv")
df.head()
```

```python
df.isnull().sum() # Перевірка на відсутні значення
df.duplicated().sum() # Перевірка на дублікати

df = df.drop_duplicates() # Видалення дублікатів (якщо є)
print(df.dtypes)

# Обробка відсутніх значень
# Заповнення середнім значенням
for col in df.select_dtypes(include = np.number).columns:  
    df[col] = df[col].fillna(df[col].mean())

df.head()
```
